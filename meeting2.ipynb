Image Blending

```python
import cv2
import matplotlib.pyplot as plt

alpha = float(input("enter alpha [0.0-1.0] : "))

img1 = cv2.imread('/content/wtc.png')
img2 = cv2.imread('/content/crashwtc.jpg')

# samakan ukuran img2 dengan ukuran img1
img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))
beta = 1.0 - alpha
blending_img = cv2.addWeighted(img1, alpha, img2, beta, 0.0)

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title("image 1")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.title("image 2")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(blending_img, cv2.COLOR_BGR2RGB))
plt.title(f"Blended(alpha={alpha:.2f})")
plt.axis("off")

plt.show()
```
