```python
# crop image[y_min:y_max, x_min:x_max]
img_crop = img[0:355, 0:400]
```

```python
import cv2
import matplotlib.pyplot as plt

# baca gambar
img = cv2.imread("/content/suwanting.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #mengubah ke RGb agar warna benar

plt.figure(figsize=(10, 5))

plt.subplot(1,2,1)
plt.imshow(img)
plt.title("original image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(img_crop)
plt.title("crop image")
plt.axis("off")

plt.show()
```
